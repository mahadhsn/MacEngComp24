```python
!pip install python-docx  # automatically installs library for user
from cryptography.fernet import Fernet
from docx import Document
import ipywidgets as widgets
from IPython.display import display
import io

# Input widgets
key_input = widgets.Text(
    description='Input your encryption key here:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='100%')
)

# Input txt file widget
file_input = widgets.FileUpload(
    description='Upload .txt file to be encrypted',
    accept='.docx',
    multiple=False,
    layout=widgets.Layout(width='50%')
)

# click-based events
output_button = widgets.Button(
    description="Encrypt",
    button_style='success'
)

# Output widget
result_output = widgets.Output()

# Define button click function
def button_clicked(b):
    with result_output:
        result_output.clear_output()

        # Retrieve key and handle encryption key generation
        key = key_input.value

        # generates a default key
        if key == "":
            key = Fernet.generate_key()
        # encrypts using the input key
        else:
            key = key.encode('utf-8')

        cipher_suite = Fernet(key)

        # Retrieve uploaded file content
        if file_input.value:
            uploaded_file = list(file_input.value.values())[0]
            file_content = uploaded_file['content']

            # Read file content and encrypt
            encrypted_message = cipher_suite.encrypt(file_content)
            encrypted_message_str = encrypted_message.decode('utf-8')
            print("Encrypted Message:", encrypted_message_str)


            # Save encrypted message to a new document
            encrypted_doc = Document()
            encrypted_doc.add_paragraph(encrypted_message_str)
            encrypted_doc.save("SECRET MESSAGE.docx")

            # Save the key to a separate document
            key_non_byte = key.decode('utf-8')
            key_doc = Document()
            key_doc.add_paragraph(key_non_byte)
            key_doc.save("KEY.docx")

            print("Encrypted message saved as 'SECRET MESSAGE.docx'")
            print("Encryption key saved as 'KEY.docx'")
        else:
            print("No file uploaded")

# Link button to function and display widgets
output_button.on_click(button_clicked)
display(key_input, file_input, output_button, result_output)
```
